```python
import networkx as nx
G = nx.complete_graph(15)
```

```python
nx.draw_networkx(G)
```

```python
sorted(nx.common_neighbors(G, 1, 2))
```

```python
preds = nx.jaccard_coefficient(G, [(0, 1), (2, 3)])
for u, v, p in preds:
    print(u, v, '->', p)
```

```python
nx.jaccard_coefficient(G)
```

```python
preds = nx.jaccard_coefficient(G,[(1,2),(5,6)])
for u, v, p in preds:
    print(u, v, '->', p)
```

```python
preds = nx.adamic_adar_index(G,[(1,2),(5,6)])
for u, v, p in preds:
    print(u, v, '->', p)
```
